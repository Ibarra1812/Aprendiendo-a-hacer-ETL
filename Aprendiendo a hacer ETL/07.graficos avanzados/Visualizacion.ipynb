#### Practicas de estilos

highlight_null: identifica y destaca elementos faltantes (valores nulos).

highlight_between: identifica y destaca elementos que están dentro de un rango específico de valores.

highlight_quantile: identifica y destaca elementos de acuerdo con el rango de cuantiles especificado.

text_gradient: funciona de manera similar a background_gradient, creando un gradiente en el texto basado en la escala numérica de los valores presentes en la columna a ser transformada o en otra.

```python
import pandas as pd
import numpy as np

datos = pd.DataFrame({'ID': [1204, 3018, 9963, 1005, 5627],
                   'Edad': [np.nan, 32, 19, np.nan, 38],
                   'Salario': [3500.00, 4200.50, 2800.00, 5700.25, 4900.75]})
```

```python
s_nan = datos.style.format({'Edad':'{:.0f}','Salario':'{:,.2f}'})\
                   .highlight_null(color='#F79747')
s_nan
```

```python
# Definir los límites superior e inferior
límite_superior = 5000.0
# El límite inferior se especifica a través de una Series que informa cuál es la columna que tendrá la aplicación de la función
límite_inferior = pd.Series([4000.0], index=['Salario'])

# Aplicar la estilización con highlight_between
s_bet = datos.style.format({'Edad':'{:.0f}','Salario':'{:,.2f}'})\
                   .highlight_between(left=límite_inferior, right=límite_superior,
                                      axis=1, color='#94B2D7')
s_bet
```

```python
df_cliente['Total']=df_cliente.sum(axis=1)
df_cliente.loc['Total']=df_cliente.sum()
df_cliente
```

```python
compra_cliente = df_cliente.style.format('{:,.2f}')
compra_cliente
```

```python
tabla={
    'selector':'td,th:not(.index_name)',
    'props':'font-weight:normal; font-family:Arial; text-align:center'
}

indice={
    'selector':'.index_name',
    'props':'font-weight:normal; font-style:italic; text-align:right; color:#696969'
}

compra_cliente.set_table_styles([tabla,indice])
```

    Mantener la simplicidad: las tablas deben ser claras y fáciles de leer. No incluyas información innecesaria o complicada.
    Título claro y conciso: es necesario insertar un título en la tabla, y debe ser claro e indicar claramente el contenido de la tabla. Evita títulos genéricos, como "Datos" o "Resultados".
    Leyenda completa: incluye una leyenda que describa todos los elementos de la tabla, como abreviaciones, unidades de medida y símbolos.
    Evitar celdas vacías: si es posible, siempre llena las celdas con información relevante.
    Destacar datos importantes: busca destacar los datos importantes dentro de la tabla, con cambios en el fondo o con texto en negrita.
    Evitar tablas muy largas: las tablas muy largas pueden ser difíciles de leer y pueden dividirse en varias tablas más pequeñas, si es posible.

```python
# Crear un DataFrame alternativo para tener un nombre de columna coherente con la visualización
df_alteracion = df.rename(columns={'pais': 'Costo por país ($)'})

# Agrupar los datos por tipo de cliente, con la categoría de productos y los países a los que estos clientes importan el producto
df_clientes = df_alteracion.pivot_table(index=['tipo_cliente', 'categoria'], columns = 'Costo por país ($)', values = 'costo_producto', aggfunc = 'sum')

# Renombrar el encabezado para nombres más coherentes
df_clientes.rename_axis(index={'tipo_cliente': 'Tipo cliente', 'categoria': 'Categoría producto'}, inplace = True)

# Agregar una columna que muestre la suma de los costos por tipo de cliente y categoría de producto
df_clientes['Total ($)'] = df_clientes.sum(axis=1)

## Crear la visualización
s_clientes = df_clientes.style.format('{:,.2f}')

# Crear los estilos base para la tabla, desde la estandarización de las fuentes, fondo blanco y alineación del encabezado en la parte superior
tabla = {
    'selector': 'td, th',
    'props': 'font-weight: normal; font-family: Arial; text-align: right; background-color: white;'
}
# Alinear verticalmente el encabezado en la parte superior permite al observador identificar dónde comienza la jerarquización de los grupos B2B y B2C
encabezado = {
    'selector': 'th',
    'props': 'vertical-align: top'
}
s_clientes.set_table_styles([tabla, encabezado])

# Agregar línea de separación en el encabezado superior
# Resaltar las categorías que generaron más costo con la aplicación de negrita en las fuentes
s_clientes.set_table_styles({
    ('B2B', 'BlueRay'): [{'selector': 'th', 'props': 'border-top: 1px solid #181818'},
              {'selector': 'td', 'props': 'border-top: 1px solid #181818'}],
    ('B2B', 'Libro'): [{'selector': 'th', 'props': 'font-weight: bold'}],
    ('B2C', 'Libro'): [{'selector': 'th', 'props': 'font-weight: bold'}],
}, overwrite=False, axis=1)

# Resaltar los elementos en cada tipo de cliente que generaron más costos con una selección
s_clientes.set_table_styles({
    'Total ($)': [{
        'selector': '.true',
        'props': 'font-weight: bold'
    }]
},overwrite=False,axis=0)

colores_columna = pd.DataFrame(['false','false','false','true','false','false','false','true'],index= df_clientes['Total ($)'].index,
                            columns = ['Total ($)'])
# Visualización final
s_clientes.set_td_classes(colores_columna)
```
