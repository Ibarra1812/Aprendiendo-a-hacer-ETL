```python
import pandas as pd
url='https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_locacao_imoveis.json'
datos=pd.read_json(url)
datos.head()
```

```python
datos_normalize=pd.json_normalize(datos['dados_locacao'])
```

```python
datos_normalize.head()
```

```python
columns=list(datos_normalize.columns)
datos_normalize=datos_normalize.explode(columns[1:])
datos_normalize.head()
```

```python
datos_normalize.info()
```

```python
datos_normalize['valor_aluguel']=datos_normalize['valor_aluguel'].apply(lambda x: x.replace('$','').replace('reais','').replace(',', '.').strip())
datos_normalize['valor_aluguel']=datos_normalize['valor_aluguel'].astype(float)
datos_normalize.head()
```

```python
datos_normalize.reset_index(drop=True,inplace=True)
datos_normalize.head()
```

```python
datos_normalize[['apartamento', 'datas_combinadas_pagamento', 'datas_de_pagamento']]=datos_normalize[['apartamento', 'datas_combinadas_pagamento', 'datas_de_pagamento']].astype(str)
datos_normalize.head()
```

```python
datos_normalize.info()
```

```python
datos_normalize['apartamento']=datos_normalize['apartamento'].apply(lambda x: x.replace('(blocoAP)','').strip())
datos_normalize.head()
```

```python
datos_normalize['datas_combinadas_pagamento'] = datos_normalize['datas_combinadas_pagamento'].dt.strftime('%Y/%m/%d')
datos_normalize['datas_de_pagamento'] = datos_normalize['datas_de_pagamento'].dt.strftime('%Y/%m/%d')
datos_normalize.head()
```

```python
datos_normalize['datas_de_pagamento'] = pd.to_datetime(datos_normalize['datas_de_pagamento'], format='%Y/%m/%d')
datos_normalize['datas_combinadas_pagamento'] = pd.to_datetime(datos_normalize['datas_combinadas_pagamento'], format='%Y/%m/%d')
datos_normalize['atraso'] = (datos_normalize['datas_de_pagamento'] - datos_normalize['datas_combinadas_pagamento']).dt.days
media_atraso = datos_normalize.groupby('apartamento')['atraso'].mean()
media_atraso.head()
```
