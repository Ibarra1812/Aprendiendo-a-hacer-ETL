```python
import pandas as pd
```

```python
url='https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_vendas_clientes.json'
datos=pd.read_json(url)
datos.head()
```

```python
datos_normalize=pd.json_normalize(datos['dados_vendas'])
```

```python
datos_normalize.head()
```

```python
columns=list(datos_normalize.columns)
datos_normalize=datos_normalize.explode(columns[1:])
datos_normalize.head()
```

```python
datos_normalize['Valor da compra']=datos_normalize['Valor da compra'].astype(str).apply(lambda x: x.replace(',','.').replace('R$','').strip())
datos_normalize['Valor da compra']=datos_normalize['Valor da compra'].astype(float)
datos_normalize.info()
```

```python
datos_normalize.reset_index(drop=True,inplace=True)
datos_normalize.head()
```

```python
datos_normalize['Cliente']=datos_normalize['Cliente'].str.lower()
datos_normalize['Cliente']=datos_normalize['Cliente'].replace(r'[^a-zA-Z\-\']',' ',regex=True).str.strip()
datos_normalize['Cliente']=datos_normalize['Cliente'].replace(r'(?<!\w)-(?!\w)',' ',regex=True).str.strip()
datos_normalize.head()
```

```python
datos_normalize.info()
```

```python
datos_normalize['Data de venda']=pd.to_datetime(datos_normalize['Data de venda'])
```

```python
datos_normalize['Fecha de venta']=datos_normalize['Data de venda'].dt.strftime('%Y/%m/%d')
datos_normalize.head()
```

```python
total_compras=datos_normalize.groupby(['Cliente'])['Valor da compra'].sum()
total_compras.head()
```
