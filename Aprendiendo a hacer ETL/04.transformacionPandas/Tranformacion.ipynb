## Datos numericos


```python
columnas=list(datos.columns)
columnas
```

```python
datos=datos.explode(columnas[3:])
#Esto es cambio de indices haciendo que los datos de la lista
#tengan sus propias filas y sean separadas en cada uno por individual

datos.head()
```

```python
datos.reset_index(drop=True, inplace=True)
#Esto se hace para que se resetee los indices de la tabla
datos.head()
```

```python
datos.info()
```

```python
import numpy as np
```

```python
datos['max_hospedes']=datos['max_hospedes'].astype(np.int64)
#Esto es para cambiar los tipos de datos object
#para interger, no sirve si en la columna tiene un valor nulo
```

```python
datos['max_hospedes']
```

```python
datos.info()
```

```python
col_numericas=['cantidad_baños','cantidad_cuartos','cantidad_camas']
datos[col_numericas]=datos[col_numericas].astype(np.int64)
```

```python
datos['evalaucion_general']=datos['evaluacion_general'].astype(np.float64)
datos.info()
```

```python
datos['precio']=datos['precio'].apply(lambda x: x.replace('$','').replace(',','').strip())
datos['precio'].head()
```

```python
datos['precio']=datos['precio'].astype(np.float64)
datos.info()
```

## DATOS DE TEXTO

```python
datos.head()
```

```python
## Tokenizacion, agarrar una palabra y contar su peso
datos['descripcion_local']=datos['descripcion_local'].str.lower()
datos.head()
```

```python
datos['descripcion_local'][3169]
```

```python
datos['descripcion_local'][3169]
```

```python
#antes de la tokenizacion tenemos la tamizacion
#que se limpian las palabras antes de tokenizarse
datos['descripcion_local']=datos['descripcion_local'].str.replace(r'[^a-zA-Z0-9\-\']',' ',regex=True)
datos['descripcion_local']=datos['descripcion_local'].str.replace(r'(?<!\w)-(?!\w)',' ',regex=True)
datos['descripcion_local'].head()
```

```python
#Luego aca se tokenizan transformandolas en vectores de str
datos['descripcion_local']=datos['descripcion_local'].str.split()
datos.head()
```

```python
datos['comodidades']=datos['comodidades'].str.replace(r'\{|}|\"','',regex=True)
```

```python
datos['comodidades']=datos['comodidades'].str.split(',')
datos.head()
```

```python
datos['descripcion_vecindad']=datos['descripcion_vecindad'].str.lower()
datos['descripcion_vecindad']=datos['descripcion_vecindad'].replace(r'[^a-zA-Z0-9\-\']',' ',regex=True)
datos['descripcion_vecindad']=datos['descripcion_vecindad'].replace(r'(?<!\w)-(?!\w)',' ',regex=True)
datos['descripcion_vecindad']=datos['descripcion_vecindad'].str.split()
datos.head()
```
