## 2. Agrupamiento de los datos

### Modificando el formato de dataFrame

para realizar las agrupamientos en los datos de emisiones, la base de datos debe estar en un formato adecuado, con los valores de emisiones todos n la misma columna. Para ello debemos modificar el formato de la base de datos para que las columnas referentes a cada año se conviertan en una unica columna

```python
emisiones_gases.loc[:,'Nível 1 - Setor':'Produto'].columns
```

```python
columnas_info=list(emisiones_gases.loc[:,'Nível 1 - Setor':'Produto'].columns)
columnas_info
```

```python
emisiones_gases.loc[:,1970:2021].columns
```

```python
columnas_emisiones=list(emisiones_gases.loc[:,1970:2021].columns)
columnas_emisiones
```

Para realizar la transformacion del dataFrame, utilizamos el metodo melt(), este metodo tiene como objetivo de transformar un dataFrame d eun formato ancho (wide) para un formato largo(wide)

```python
emisiones_gases.melt(id_vars=columnas_info, value_vars=columnas_emisiones, var_name='Año', value_name='Emisiones')
#id_vars van las columnas que no queremos que se modifique
#value_vars las que si
#var_name nombre de la nueva columna de variables
#value_name nombre de la colum valores de estas variables nuevas
```

### Analisis de las emisiones
vamos a realizar un analisi, pero esta vez, de la cantidad de la emision total por tipo de gas, para esto, vamos a realizar un agrupamiento de los datos con la funcion de groupby()

```python
emisiones_por_año=emisiones_gases.melt(id_vars=columnas_info, value_vars=columnas_emisiones, var_name='Año', value_name='Emisiones')
```

```python
emisiones_por_año.groupby("Gás").groups
#Te devuelve un diccionario con los tipos de gas
# Entonces devuelve valores segun los tipos dado
```

Podemos revisar que los datos fueron agrupados en cada grupo con el uso del atributo groups y realizar selecciones de grupos especificos con el metodo get_group()

```python
emisiones_por_año.groupby("Gás").get_group("CO2 (t)")
#Nos da solo los valores de CO2
#El get_group accede a todos los valores CO2 (t)
# Y los enlista
```

Los metodos de agregacion posibles con la utilizacion de groupby() son los siguientes


*   count() conteo

```python
#Aca estamos agrupando a todos los co2
#tomando  los valores co2 declarados primero
#luego la posicion de la columna solo usando el de la posicion 0
gas_por_sector.xs('CO2 (t)', level=0)
```

## Poblacion de los estados

Importaremos una nueva base de datos con informacion de la poblacion de cada estado de Brasil, para luego encontrar la emision per capita de cada estado
Los datos se obtuvieron a traves del censo del IBGE

```python
poblacion_estados=pd.read_excel('/content/drive/MyDrive/Colab_Notebooks/2045-ProyectoFinal/2045-ProyectoFinal/data/POP2022_Municipios.xls', header = 1, skipfooter = 34)
poblacion_estados
```

Como la informacion se refiere a municipios, realizaremos una agregacvion basada en la columna estado

```python
poblacion_estados.groupby('UF').sum(numeric_only=True)
```

Notamos que la informacion de la poblacion no estaba agregada, por lo que sera necesario realizar algun tratamiento en esta columna para devolver los datos correctamente.

```python
#poblacion_estados['POPULAÇÃO'].astype(int)
```

```python
poblacion_estados[poblacion_estados['POPULAÇÃO'].str.contains('\(', na = False)]
```

```python
poblacion_estados = poblacion_estados.assign(poblacion_sin_parentesis=poblacion_estados['POPULAÇÃO'].astype(str).str.replace(r'\(\d{1,2}\)','',regex=True),
                      poblacion=lambda x: x.loc[:,'poblacion_sin_parentesis'].str.replace(r'\.','',regex=True))
```

```python
poblacion_estados[poblacion_estados['POPULAÇÃO'].str.contains('\(', na = False)]
```

```python
poblacion_estados.loc[:,'poblacion']= poblacion_estados['poblacion'].astype(int)
poblacion_estados
```

```python
# Re-load the original data to ensure 'POPULAÇÃO' column is present
poblacion_estados = pd.read_excel('/content/drive/MyDrive/Colab_Notebooks/2045-ProyectoFinal/2045-ProyectoFinal/data/POP2022_Municipios.xls', header = 1, skipfooter = 34)

# Clean and convert the 'POPULAÇÃO' column to an integer 'poblacion' column
poblacion_estados = poblacion_estados.assign(
    poblacion_sin_parentesis=poblacion_estados['POPULAÇÃO'].astype(str).str.replace(r'\(\d{1,2}\)', '', regex=True)
)
poblacion_estados.loc[:, 'poblacion'] = poblacion_estados['poblacion_sin_parentesis'].str.replace(r'\.','',regex=True).astype(int)

# Now perform the grouping and sum
poblacion_estados = poblacion_estados.groupby('UF')['poblacion'].sum().reset_index()
poblacion_estados
```

# Uniendo datos

Con la informacion procesada podemos unir los datos de emisiones con los datos de la poblacion, itilizando merge().

```python
emision_estados= emisiones_por_año[emisiones_por_año['Año'] == 2021].groupby('Estado').sum(numeric_only=True).reset_index()
emision_estados
```

```python
datos_agrupados= pd.merge(emision_estados, poblacion_estados, left_on='Estado', right_on='UF')
datos_agrupados
```

Comprobemos si hay algun indicio de si las emisiones aumentan co nel tamano de la poblacion

```python
datos_agrupados.plot(kind='scatter', x='poblacion', y='Emisiones', figsize=(8,6))
```

Si queremos conocer la informacion de cada estado en lso puntos dle grafico, podemos utilizar otra biblioteca para obtener un grafico mas elaborado

```python
import plotly.express as px
```

```python
px.scatter(data_frame = datos_agrupados, x = 'poblacion', y = 'Emisiones', text = 'Estado', opacity = 0.8)
```

```python
datos_agrupados = datos_agrupados.assign(emision_per_capita = datos_agrupados['Emisiones']/datos_agrupados['poblacion']).sort_values('emision_per_capita', ascending = False)
datos_agrupados
```

```python
px.bar(data_frame= datos_agrupados, x= 'Estado', y= 'emision_per_capita')
```
