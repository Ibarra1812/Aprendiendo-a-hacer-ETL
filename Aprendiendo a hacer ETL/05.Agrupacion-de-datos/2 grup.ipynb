```python
# En regex cuando usamos \ es para buscar el caracter siguiente
poblacion_estados['POPULAÇÃO'].str.contains('\(',na=False)
```

```python
# muestra a todos los true
poblacion_estados[poblacion_estados['POPULAÇÃO'].str.contains('\(',na=False)]
```

```python
poblacion_estados['POPULAÇÃO'].isnull().sum()
```

```python
poblacion_estados = poblacion_estados.assign(poblacion_sin_parentesis=poblacion_estados['POPULAÇÃO'].astype(str).str.replace(r'\(\d{1,2}\)','',regex=True),
                      poblacion=lambda x: x.loc[:,'poblacion_sin_parentesis'].str.replace(r'\.','',regex=True))
```

```python
poblacion_estados.sample(10)
```

```python
poblacion_estados[poblacion_estados['POPULAÇÃO'].str.contains('\(',na=False)]
```

```python
poblacion_estados['poblacion'].isnull().sum()
```

```python
poblacion_estados['poblacion'] = poblacion_estados['poblacion'].fillna(0).astype('int64')
```

```python
poblacion_estados.groupby('UF')[['poblacion']].sum(numeric_only=True).reset_index()
```
