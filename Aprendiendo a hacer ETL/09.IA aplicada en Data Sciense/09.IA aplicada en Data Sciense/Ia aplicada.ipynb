### **2da Pregunta:** ¿Cuál fue la facturación por categoria en 2023?

**Observación:** Para la facturación, consideraremos aquí el producto entre el precio unitario y la cantidad, sumados al costo de envío de cada producto.


> #### <strong><font color="7b71ff" size=6>Prompt:</font></strong>

Tenemos un DataFrame llamado "df" y queremos construir un gráfico de barras horizontales que muestre la facturación por categoría de Zoop.

En "df", tenemos las siguientes columnas:
* categoria: categorías de los productos
* precio_unitario: precio de una unidad del producto
* cantidad: cantidad vendida del producto
* envio: costo de envío total de la compra.

Deseamos crear una visualización que muestre la facturación de cada categoría, donde la facturación se calcula multiplicando el precio unitario por la cantidad y sumando el costo de envío.

Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.

````python

# Calculamos la facturación total por cada fila
# La fórmula utilizada es: $Facturación = (precio\_unitario \times cantidad) + envio$
df_proyecto['facturacion'] = (df_proyecto['precio_unitario'] * df_proyecto['cantidad']) + df_proyecto['envio']

# Agrupamos los datos por categoría, sumamos la facturación y ordenamos de mayor a menor
facturacion_por_categoria = df_proyecto.groupby('categoria')['facturacion'].sum().sort_values(ascending=False)

# Configuramos el estilo visual
sns.set_theme(style="whitegrid")

# Creamos el gráfico de barras horizontales
sns.barplot(
    x=facturacion_por_categoria.values,
    y=facturacion_por_categoria.index,
    palette='viridis'
)

# Personalizamos el gráfico con títulos y etiquetas claras
plt.title('Facturación Total por Categoría', fontsize=15)
plt.xlabel('Facturación (Suma de ventas + envíos)', fontsize=12)
plt.ylabel('Categoría de Producto', fontsize=12)

# Ajustamos para que las etiquetas no se corten
plt.tight_layout()
````

### **3a Pregunta:** ¿Cómo fueron las ventas a lo largo de los meses en 2023?
> #### <strong><font color="7b71ff" size=6>Prompt:</font></strong>

Tenemos un DataFrame llamado "df" y queremos construir un gráfico de líneas que muestre las ventas totales mensuales de Zoop.

En "df", tenemos las siguientes columnas:
* fecha: con las fechas de las compras en formato datetime (`aaaa-mm-dd`)
* facturación: calculada como el producto del precio unitario y la cantidad, sumado al costo de envío de cada venta.

Deseamos crear una visualización que muestre las ventas por mes. Primero, agrupa los datos por mes y luego crea una nueva columna llamada "mes" que contenga el nombre de cada mes, traduciendo los meses del índice mediante el siguiente diccionario. Utiliza la columna "fecha" para agrupar los datos y construir el gráfico. La columna "mes" solo debe usarse para cambiar las etiquetas del eje x.

```
meses = {
    'January': 'Ene',
    'February': 'Feb',
    'March': 'Mar',
    'April': 'Abr',
    'May': 'May',
    'June': 'Jun',
    'July': 'Jul',
    'August': 'Ago',
    'September': 'Sep',
    'October': 'Oct',
    'November': 'Nov',
    'December': 'Dic'
}
```


Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.


````python

# Diccionario para traducir los meses
meses = {
    'January': 'Ene',
    'February': 'Feb',
    'March': 'Mar',
    'April': 'Abr',
    'May': 'May',
    'June': 'Jun',
    'July': 'Jul',
    'August': 'Ago',
    'September': 'Sep',
    'October': 'Oct',
    'November': 'Nov',
    'December': 'Dic'
}

# Agrupamos por mes y sumamos la facturación
# Usamos resample con 'ME' (Month End) para agrupar por el final del mes
ventas_mensuales = df_proyecto.resample('ME', on='fecha')['facturacion'].sum().reset_index()

# Creamos la columna 'mes' extrayendo el nombre del mes en inglés y mapeándolo al diccionario
ventas_mensuales['mes'] = ventas_mensuales['fecha'].dt.month_name().map(meses)

# Configuramos el estilo del gráfico
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 6))

# Creamos el gráfico de líneas
# Usamos marcador 'o' para señalar los puntos de cada mes
sns.lineplot(data=ventas_mensuales, x='mes', y='facturacion', marker='o', linewidth=2.5, color='royalblue')

# Personalizamos títulos y ejes
plt.title('Evolución de la Facturación Mensual - 2023', fontsize=16)
plt.xlabel('Mes', fontsize=12)
plt.ylabel('Facturación Total', fontsize=12)

# Ajustamos los márgenes para que no se corte nada
plt.tight_layout()
````

* facturacion: con los valores de las ventas
* metodo_pago: con el método de pago elegido en la compra

Deseamos crear una visualización que muestre las ventas por trimestre, agrupando los métodos de pago de cada trimestre correspondientes a cada compra.

Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.

````python


# Creamos una columna para el trimestre (Q1, Q2, Q3, Q4)(Q es una variable predeterminada viene de quarter)
df_proyecto['trimestre'] = df_proyecto['fecha'].dt.to_period('Q').astype(str)

# Agrupamos por trimestre y método de pago, sumando la facturación
ventas_trimestrales = df_proyecto.groupby(['trimestre', 'metodo_pago'])['facturacion'].sum().unstack()
#un stack hace que no sea multi index la tabla sino que sea que cada fila del indice multiple se reparta por las columnas
# Generamos el gráfico de barras apiladas
# El parámetro stacked=True es el que permite ver la distribución interna por trimestre
ventas_trimestrales.plot(kind='bar', stacked=True, figsize=(12, 7), colormap='tab10')

# Personalizamos el gráfico para que sea legible
plt.title('Distribución de Ventas por Trimestre y Método de Pago', fontsize=16)
plt.xlabel('Trimestre', fontsize=12)
plt.ylabel('Facturación Total', fontsize=12)
plt.xticks(rotation=0) # Mantenemos las etiquetas de los trimestres horizontales
plt.legend(title='Método de Pago', bbox_to_anchor=(1.05, 1), loc='upper left')

# Ajustamos el diseño final
plt.tight_layout()
````

Ahora ha llegado el momento del gráfico de sectores. Ya está bien encaminado, pero necesitamos realizar algunos ajustes.

````python
def proporcion_cashback():
  # 1. Preparamos los datos del conteo
  conteo_cashback = df_proyecto['cashback'].value_counts()

  # 2. Configuramos la figura y el fondo del gráfico
  fig, ax = plt.subplots(figsize=(6, 6))
  fig.set_facecolor(GRIS_1)
  ax.set_facecolor(GRIS_1)

  # 3. Definimos el mapeo de colores según las porciones
  # 'Sí' va con AZUL_1 y 'No' con GRIS_3
  colores_porciones = [AZUL_1 if label == 'Sí' else GRIS_3 for label in conteo_cashback.index]

  # 4. Creamos el gráfico de dona
  # Capturamos patches, texts (labels externos) y autotexts (porcentajes internos)
  patches, texts, autotexts = ax.pie(
      conteo_cashback,
      labels=conteo_cashback.index,
      colors=colores_porciones,
      autopct='%1.1f%%',
      pctdistance=0.8,
      startangle=180,
      wedgeprops={'width': 0.4} # Establece el grosor de la dona
  )

  # 5. Personalizamos las etiquetas externas (Si/No)
  # Deben tener el mismo color que su porción y tamaño 12
  for i, text in enumerate(texts):
      text.set_color(colores_porciones[i])
      text.set_size(12)

  # 6. Personalizamos los porcentajes internos
  # Deben ser blancos y tamaño 12
  for autotext in autotexts:
      autotext.set_color(BLANCO)
      autotext.set_size(12)

  # 7. Título y ajustes finales
  plt.title('¿Participan en el programa de cashback?', fontsize=18, color=GRIS_5, pad=20)

  # Aseguramos que el gráfico sea circular
  ax.axis('equal')
  plt.tight_layout()
  return plt
````

````python
def box_plot_edad_sexo():
  # 1. Configuración de la figura y el fondo
  fig, ax = plt.subplots(figsize=(10, 6))
  fig.set_facecolor(GRIS_1)
  ax.set_facecolor(GRIS_1)

  # 2. Creación del boxplot con la paleta de Zoop
  sns.boxplot(
      x='sexo_biologico',
      y='edad',
      data=df_proyecto,
      palette=[AZUL_1, AQUA_1]
  )

  # 3. Personalización del Título
  plt.title('Distribución de la Edad por Sexo Biológico de los clientes', fontsize=18, color=GRIS_5, loc='left', pad=25)

  # 4. Ajuste de ejes y etiquetas
  plt.xlabel('') # Suprimimos la etiqueta del eje x
  plt.ylabel('Edad', fontsize=12, color=AZUL_1)
  plt.xticks(fontsize=12, color=AZUL_1)
  plt.yticks(fontsize=12, color=AZUL_1)

  # 5. Configuración de la cuadrícula y eliminación de bordes
  ax.grid(axis='y', linestyle='-', alpha=0.3)
  ax.set_axisbelow(True) # El grid queda detrás del gráfico

  # Suprimimos todos los bordes
  for spine in ax.spines.values():
      spine.set_visible(False)

  # 6. Preparación de datos para las anotaciones
  # Calculamos las métricas agrupadas para los dos grupos
  stats = df_proyecto.groupby("sexo_biologico")["edad"].describe()
  # También necesitamos los cuartiles específicos
  q1 = df_proyecto.groupby("sexo_biologico")["edad"].quantile(.25)
  q3 = df_proyecto.groupby("sexo_biologico")["edad"].quantile(.75)
  mediana = df_proyecto.groupby("sexo_biologico")["edad"].median()

  # Anotación para el primer grupo (índice 1 en la lógica del prompt)
  texto_1 = (
      f"Mínimo = {stats['min'].iloc[1]}\n"
      f" 1º quartil = {q1.iloc[1]}\n"
      f" 2º quartil = {mediana.iloc[1]}\n"
      f" 3º quartil = {q3.iloc[1]}\n"
      f" Máximo = {stats['max'].iloc[1]}"
  )

  ax.annotate(
      texto_1,
      xy=(-0.1, 45),
      xytext=(-0.4, 50),
      color=AZUL_1,
      fontsize=10,
      bbox=dict(boxstyle='round,pad=0.5', fc=BLANCO, ec=GRIS_3),
      arrowprops=dict(arrowstyle='->', color=AZUL_1, linewidth=1.5)
  )

  # Anotación para el segundo grupo (índice 0 en la lógica del prompt)
  texto_2 = (
      f"Mínimo = {stats['min'].iloc[0]}\n"
      f" 1º quartil = {q1.iloc[0]}\n"
      f" 2º quartil = {mediana.iloc[0]}\n"
      f" 3º quartil = {q3.iloc[0]}\n"
      f" Máximo = {stats['max'].iloc[0]}"
  )

  ax.annotate(
      texto_2,
      xy=(0.9, 45),
      xytext=(0.6, 50),
      color=AZUL_1,
      fontsize=10,
      bbox=dict(boxstyle='round,pad=0.5', fc=BLANCO, ec=GRIS_3),
      arrowprops=dict(arrowstyle='->', color=AZUL_1, linewidth=1.5)
  )

  plt.tight_layout()
  return plt
````
